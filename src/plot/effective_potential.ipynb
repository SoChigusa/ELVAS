# Plot the RG flow and effective potential

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the subplots
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8,6))

# Load data
data = np.loadtxt('../output/2306_RGEFlow.dat', skiprows=1, usecols=range(2))
mu = data[:,0]
lam = data[:,1]
V = 0.25 * lam * np.power(mu, 4)

# Plot the data on the subplots
ax1.plot(mu, lam, 'm')
ax2.plot(mu, V, 'b')

# Auxiliary lines
zero = [0] * mu.size
ax1.plot(mu, zero, 'black', lw=1, ls=':')

# Set scales and ranges
ax2.set_xscale('log')
ax2.set_ylim([-1e36,1e36])
ax2.set_xlim([1e2, 1e20])

# Add labels and titles
ax1.set_ylabel(r'$\lambda(h)$', size=20)
ax2.set_ylabel(r'$V_{\rm eff}(h)$', size=20)
ax2.set_xlabel(r'$h\, /\, {\rm GeV}$', size=20)
for ax in [ax1, ax2]:
  ax.tick_params(labelsize=15)

# Adjust spacing between subplots
plt.subplots_adjust(hspace=0)

# Save the plot
plt.savefig('effective_potential.pdf', bbox_inches='tight')
```
